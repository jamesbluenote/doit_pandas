### - (p.125) 데이터 연결 응용
#### (1) 인덱스 새로 지정 : ignore_index = True

```python
row_concat_i = pd.concat([df1, df2, df3], ignore_index=True)
row_concat_i
```

#### (2) 열 방향으로 데이터 연결 : axis = 1

```python
col_concat = pd.concat([df1, df2, df3], axis = 1)
col_concat
```

```python
# 열 이름이 동일한 열이 있는 경우 같은 열 이름 모두 추출
col_concat['A']
```

#### (3) 신규 열 추가

```python
col_concat['new_col_list'] = ['n1', 'n2', 'n3', 'n4']
col_concat
```

#### (4) 열 방향 연결 & 인덱스 새로 지정 

```python
pd.concat([df1, df2, df3], axis=1, ignore_index=True)
```

#### (5) 열 이름이 다를 경우, 연결하면 NaN 처리됨

```python
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']
```

```python
row_concat = pd.concat([df1, df2, df3], ignore_index = True)
row_concat
```

#### (6) 공통 열만 연결 : join = 'inner'

```python
pd.concat([df1, df2, df3], ignore_index = False, join = 'inner')
```

```python
pd.concat([df1, df3], ignore_index = False, join = 'inner')
```

#### (7) 행 번호가 다를 경우, 연결하면 NaN 처리됨

```python
df1.index = [0, 1, 2, 3]
df2.index = [4, 5, 6, 7]
df3.index = [0, 2, 5, 7]
```

```python
df3
```

#### (8) 공통 행 번호만 연결 : axis = 1, join = 'inner'

```python
pd.concat([df1, df3], axis = 1, join = 'inner')
```

### - (p.132) left_df.merge(right_df, left_on=, right_on=)

```python
person = pd.read_csv('survey_person.csv')
site = pd.read_csv('survey_site.csv')
survey = pd.read_csv('survey_survey.csv')
visited = pd.read_csv('survey_visited.csv')
```

```python
# for x, y in enumerate(): 연습
survey_list = ['survey_person', 'survey_site', 'survey_survey', 'survey_visited']
filename_list = []

for idx, filename in enumerate(survey_list):
    filename_list.append(filename + '.csv')

person = pd.read_csv(filename_list[0])
site = pd.read_csv(filename_list[1])
survey = pd.read_csv(filename_list[2])
visited = pd.read_csv(filename_list[3])
```

```python
person
```

```python
survey_shapes = {'person' : person.shape, 
                 'site'   : site.shape, 
                 'survey' : survey.shape, 
                 'visited': visited.shape}
survey_shapes
```

```python
visited_subset = visited.loc[[0, 2, 6], ]
```

```python
visited_subset
```

```python
site
```

#### * merge : 기본적으로 내부 조인(inner join)

```python
o2o_merge = site.merge(visited_subset, left_on='name', right_on='site')
o2o_merge
```

```python
m2o_merge = site.merge(visited, left_on='name', right_on='site')
m2o_merge
```

```python
ps = person.merge(survey, left_on = 'ident', right_on = 'person')
ps
```

```python
vs = visited.merge(survey, left_on = 'ident', right_on = 'taken')
vs
```

#### * 복수 개의 기준열 조인

```python
ps_vs = ps.merge(vs, 
                 left_on=['ident', 'taken', 'quant', 'reading'], 
                 right_on=['person', 'ident', 'quant', 'reading'])
ps_vs
```
