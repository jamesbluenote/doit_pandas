## 07-2. 열 이름 관리

```python
ebola_raw = pd.read_csv('country_timeseries.csv')
ebola = ebola_raw.copy()
ebola.head()
```

```python
ebola.iloc[:5, [0, 1, 2, 3, 10, 11]]
```

```python
ebola_long = pd.melt(ebola, id_vars = ['Date', 'Day'])
```

```python
ebola_long.head()
```

### - (p.159) df.str.split() 메서드로 열 이름 분리

#### * Series 로 출력

```python
# (비추) : df.variable --> 변수명(column name)인지 메서드/어트리뷰트인지 구분이 안됨
# 아래 방식으로 해라 : df['variable']
variable_split = ebola_long.variable.str.split('_')
variable_split
```

```python
# 열 이름을 df.['col_name'] 로 지정해라. 앞의 방식은 variable 이 변수명인지, 메서드 또는 어트리뷰트인지 헷갈린다 
variable_split = ebola_long['variable'].str.split('_')
```

```python
variable_split.head()
```

```python
type(variable_split)
```

```python
type(variable_split[0])
```

#### * DataFrame 으로 출력

```python
variable_split_df = ebola_long['variable'].str.split('_', expand = True)
variable_split_df.head()
```

```python
type(variable_split_df)
```

```python
status_values = variable_split.str.get(0)
[status_values.unique(), status_values.nunique()]
```

```python
country_values = variable_split.str.get(1)
[country_values.unique(), country_values.nunique()]
```

### - (p.161) 분리한 열 이름 df에 추가

```python
ebola_long.head()
```

```python
ebola_long['status'] = status_values
ebola_long['country'] = country_values
```

```python
ebola_long.head()
```

#### * df.concat([ ]) 메서드로 df에 열 추가

```python
variable_split = ebola_long['variable'].str.split('_', expand = True)
variable_split.head()
```

```python
variable_split.columns = ['status2', 'country2']
```

```python
variable_split.head()
```

```python
ebola_parsed = pd.concat([ebola_long, variable_split], axis = 1)
```

```python
ebola_parsed
```

## 07-3. 여러 열을 하나로 정리

```python
weather_raw = pd.read_csv('weather.csv')
weather = weather_raw.copy()
weather.head()
```

```python
weather.shape
```

### - (p.163) df.melt() 로 피벗

```python
weather_melt = weather.melt(id_vars = ['id', 'year', 'month', 'element'],
                            var_name = 'day', value_name = 'temp')
```

```python
weather_melt
```

### - (p.164) df.pivot_table() 메서드로 행/열 위치 변경

```python
weather_tidy = weather_melt.pivot_table(index = ['id', 'year', 'month', 'day'],
                                        columns = 'element', 
                                        values = 'temp')
```

```python
weather_tidy
```

#### * Null 값도 포함 (dropna = False)

```python
weather_tidy_all = weather_melt.pivot_table(index = ['id', 'year', 'month', 'day'],
                                            columns = 'element',
                                            values = 'temp',
                                            dropna = False)
weather_tidy_all
```

#### * 인덱스 새로 지정 : df.reset_index()

```python
weather_tidy.reset_index()
```

```python
weather_tidy_all.reset_index()
```

## 07-4. 중복 데이터

### - (p.167) 중복 데이터 처리

```python
billboard = billboard_raw.copy()
billboard.head()
```

```python
billboard_long = billboard.melt(id_vars = ['year', 'artist', 'track', 'time', 'date.entered'],
                                var_name = 'week', 
                                value_name = 'rating')
```

```python
billboard_long.shape
```

```python
billboard_long.head()
```

#### (1) 데이터 확인

```python
billboard_long[billboard_long['track'] == 'Loser']
```

#### (2) 중복 데이터 열만 따로 저장

```python
billboard_songs = billboard_long[['year', 'artist', 'track', 'time']]
billboard_songs.shape
```

#### (3) 중복 데이터 제거 : df.drop_duplicates()

```python
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs.shape
```

#### (4) 중복 제거한 데이터프레임에 id 추가 : range(len(df))

```python
billboard_songs['id'] = range(len(billboard_songs))
```

```python
billboard_songs.head()
```

#### (5) 데이터 합치기 : df.merge()

```python
billboard_ratings = billboard_long.merge(billboard_songs, on = ['year', 'artist', 'track', 'time'])
```

```python
billboard_ratings.shape
```

```python
billboard_ratings.head()
```

## 07-5. 대용량 데이터 처리

### - (p.169) 여러 개로 나누어진 데이터 불러오기

```python
import urllib.request
```

```python
urllib.request.urlretrieve?
```

```python
with open('raw_data_urls.txt', 'r') as data_urls:
    for line, url in enumerate(data_urls):
        if line == 5:
            break
        fn = url.split('/')[-1].strip()
        fp = os.path.join('', '..\data', fn)
        print(url)
        print(fp)
        urllib.request.urlretrieve(url, fp)
```
